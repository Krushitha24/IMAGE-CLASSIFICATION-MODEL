```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale=1./255,shear_range = 0.2,zoom_range=0.2,horizontal_flip=True)
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
x_train = train_datagen.flow_from_directory(r'C:\Users\krush\Downloads\codetech_ml\codetech task-3\training_set\training_set',target_size=(64,64),batch_size=10,class_mode="binary")
x_test = test_datagen.flow_from_directory(r"C:\Users\krush\Downloads\codetech_ml\codetech task-3\test_set\test_set",target_size=(64,64),batch_size=10,class_mode="binary")
```

```python
x_train.class_indices
```

```python
x_test.class_indices
```

```python
#CNN model
```

```python
model=Sequential()
```

```python
model.add(Convolution2D(32,(3,3),input_shape=(64,64,3),activation='relu'))
```

```python
model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
model.add(Convolution2D(32,(3,3),activation='relu'))
```

```python
model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
model.add(Flatten())
```

```python
#Adding the dense layers
```

```python
model.add(Dense(units=128,kernel_initializer='random_uniform',activation='relu'))
```

```python
model.add(Dense(units=128,kernel_initializer='random_uniform',activation='relu'))
```

```python
model.add(Dense(units=128,kernel_initializer="random_uniform",activation='relu'))
```

```python
model.add(Dense(units=128,kernel_initializer='random_uniform',activation='relu'))
```

```python
model.add(Dense(units=128,kernel_initializer='random_uniform',activation='relu'))
```

```python
model.add(Dense(units=1,kernel_initializer='random_uniform',activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
steps_per_epoch = len(x_train)
validation_steps = len(x_test)
```

```python
model.fit(
    x_train,
    steps_per_epoch=steps_per_epoch,
    epochs=5,
    validation_data=x_test,
    validation_steps=validation_steps
)
```

```python
_, accuracy = model.evaluate(x_train)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
from tensorflow.keras.preprocessing import image
import numpy as np

# Load image
img_path = r'C:\Users\krush\Downloads\codetech_ml\codetech task-3\test_set\test_set\cats\cat.4993.jpg'
img = image.load_img(img_path, target_size=(64, 64))  

# Convert to array
img_array = image.img_to_array(img)

# Expand dimensions to simulate batch size
img_array = np.expand_dims(img_array, axis=0)

# Normalize 
img_array = img_array / 255.0
```

```python
prediction = model.predict(img_array)

# For binary classification (sigmoid output)
if prediction[0][0] > 0.5:
    print("Predicted: Dog")
else:
    print("Predicted: Cat")
```
